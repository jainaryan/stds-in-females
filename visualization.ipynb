```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('results.csv', header =None)
df
```

```python
df.columns = ['models', 'f1-score', 'precision', 'recall', 'accuracy']
df.drop(df.index[0], inplace=True)
df
```

```python

df['models'] = df['models'].astype(str)
df['f1-score'] = pd.to_numeric(df['f1-score'], errors='coerce')

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['models'], df['f1-score'], color='red')
plt.xlabel('Models')
plt.ylabel('F1-Score')
plt.title('Model F1-Scores')
plt.xticks(rotation=45, ha='right')

# Specify y-ticks
plt.yticks(np.arange(0, df['f1-score'].max(), step=0.1))

plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Convert 'models' to string and 'recall' to numeric
df['models'] = df['models'].astype(str)
df['recall'] = pd.to_numeric(df['recall'], errors='coerce')

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['models'], df['recall'], color='blue')  # Change 'f1-score' to 'recall' and color to 'blue'
plt.xlabel('Models')
plt.ylabel('Recall')  # Change y-label to 'Recall'
plt.title('Model Recall')
plt.xticks(rotation=45, ha='right')

# Specify y-ticks
plt.yticks(np.arange(0, df['recall'].max(), step=0.1))  # Change 'f1-score' to 'recall'

plt.tight_layout()

# Show the plot
plt.show()
```

```python

df['models'] = df['models'].astype(str)
df['precision'] = pd.to_numeric(df['precision'], errors='coerce')

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['models'], df['precision'], color='green') 
plt.xlabel('Models')
plt.ylabel('Precision')  
plt.title('Model Precision')
plt.xticks(rotation=45, ha='right')

# Specify y-ticks
plt.yticks(np.arange(0, df['precision'].max(), step=0.1)) 

plt.tight_layout()

# Show the plot
plt.show()
```

```python

df['models'] = df['models'].astype(str)
df['recall'] = pd.to_numeric(df['accuracy'], errors='coerce')

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['models'], df['accuracy'], color='blue')  
plt.xlabel('Models')
plt.ylabel('Accuracy') 
plt.title('Model Accuracy')
plt.xticks(rotation=45, ha='right')

# Specify y-ticks
plt.yticks(np.arange(0, df['accuracy'].max(), step=0.1)) 

plt.tight_layout()

# Show the plot
plt.show()
```
