```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data.csv', header =None)
df
```

```python
df.columns = ['Models', 'Score']
df
```

```python

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['Models'], df['Score'], color='red')
plt.xlabel('Models')
plt.ylabel('Score')
plt.title('Model Scores')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```
