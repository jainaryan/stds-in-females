```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
file_path = './data./Data Set for Chapter - Sheet1.csv'
df = pd.read_csv(file_path)
df = df.drop('S.no',axis = 1)
df = df.dropna()
df
```

```python
std_positive = df[df['STD Status'] == 1]
std_negative = df[df['STD Status'] == 0]
```

```python
print("number of positive samples: ",len(std_positive))
print("number of negative samples: ",len(std_negative))
```

```python
# Assuming len(std_positive) and len(std_negative) are your values
positive_samples = len(std_positive)
negative_samples = len(std_negative)

# Creating a horizontal bar plot
plt.barh(['Infected Samples', 'Uninfected Samples'], [positive_samples, negative_samples], color=['green', 'red'], height = 0.8)

# Adding labels and title
plt.xlabel('Number of Samples')

# Display the plot
plt.show()
```

```python
df.rename(columns = { 'Protection Usage (0: Never, 1: Sometimes, 2: Always)':'Protection Usage'}, inplace = True) 
```

```python
column_array = []
for col in df.columns:
    column_array.append(col)
# Create a subset of the dataframe with selected columns
subset_df = df[column_array]

# Compute the correlation matrix
correlation_matrix = subset_df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(5, 5))

# Create a heatmap using seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)

# Display the plot
plt.show()
```

```python
binary_plot_columns = ['Age', 'Intimate Partners', 'Education', 'Protection Usage (0: Never, 1: Sometimes, 2: Always)', 'Symptoms']

# Create a subset of the dataframe with selected columns
binary_plot_df = df[binary_plot_columns]

# Set up the pair plot
sns.set(style="ticks", color_codes=True)
pair_plot = sns.pairplot(binary_plot_df, hue="Symptoms", markers=["o", "s"])

# Add a title to the pair plot
pair_plot.fig.suptitle('Binary Plots of Age, Intimate Partners, Education, Protection Usage, and Symptoms', y=1.02)

# Display the plot
plt.show()
```

```python


# Bar chart for comparing the number of intimate partners
plt.figure(figsize=(12, 6))
sns.barplot(x='Symptoms', y='Intimate Partners', data=df, ci=None)
plt.title('Comparison of Intimate Partners among Individuals with and without STD Symptoms')
plt.xlabel('STD Symptoms (0: No Symptoms, 1: Symptoms)')
plt.ylabel('Number of Intimate Partners')
plt.show()

# Stacked bar chart for distribution of protection usage
plt.figure(figsize=(12, 6))
sns.countplot(x='Intimate Partners', hue='Protection Usage (0: Never, 1: Sometimes, 2: Always)', data=df)
plt.title('Distribution of Protection Usage for Different Levels of Intimate Partners')
plt.xlabel('Number of Intimate Partners')
plt.ylabel('Count')
plt.legend(title='Protection Usage (0: Never, 1: Sometimes, 2: Always)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='STD Status', hue='Protection Usage (0: Never, 1: Sometimes, 2: Always)', data=df)
plt.title('Distribution of Protection Usage based on STD Status')
plt.xlabel('STD Status (0: No STD, 1: STD)')
plt.ylabel('Count')
plt.legend(title='Protection Usage (0: Never, 1: Sometimes, 2: Always)')
plt.show()
```
